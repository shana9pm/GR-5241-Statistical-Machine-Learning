```python
import matplotlib.pyplot as plt
%matplotlib inline
import ast
import numpy as np
```

```python
with open("final_result.txt","r") as f:
    raw=f.readline()
```

```python
mylist = ast.literal_eval(raw)
allPars,alpha,val_acc,test_acc=mylist
```

```python
val_err=np.array(val_acc)
test_err=np.array(test_acc)
```

```python
thred=20
iter=np.arange(len(val_acc))[0:thred]
plt.plot(iter,val_err[0:thred],color='r',label='Cross-Val error')
plt.plot(iter,test_err[0:thred],color='blue',label='Test error')
plt.xlabel("Iterations")
plt.ylabel("Error Rate")
plt.legend(loc='upper right')
plt.show()
```
